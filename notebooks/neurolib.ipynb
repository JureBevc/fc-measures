```python
import neurolib
from neurolib.models.aln import ALNModel
from neurolib.utils.loadData import Dataset
import matplotlib.pyplot as plt
```

```python
model = ALNModel()
model.params['sigma_ou'] = 0.1 # add some noise

model.run()
```

```python
plt.plot(model.t, model.output.T)
```
